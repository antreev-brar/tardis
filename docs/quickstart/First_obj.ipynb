```python
from tardis import run_tardis
from tardis.io.atom_data.util import download_atom_data
```

```python
# the data is automatically downloaded
download_atom_data('kurucz_cd23_chianti_H_He')
```

```python
sim = run_tardis('first_obj_yml.yml')
```

```python
%pylab inline
#if pandas is not already installed run this command
#!pip install pandas
import pandas 
import matplotlib.pyplot as plt

figure(figsize=(20,6))
plotdf = pandas.DataFrame(sim.model._abundance.T) 
elements=plotdf.columns.values
plotdf['velocity'] = sim.model.v_inner.cgs

#print(sim.model._abundance)
#print(type(sim.model._abundance.__dict__))
#print(type(elements))
#print(plotdf.columns.values)
print(plotdf)
print(elements[:])


for atomic_number in elements:
    plt.plot( plotdf['velocity'], plotdf[atomic_number],marker='o', markersize=12, linewidth=4,label=atomic_number)

plt.title("Abundance vs Velocity", fontsize=18)
plt.ylabel('Abundance', fontsize=18)
plt.xlabel("Velocity(Inner shell boundary)", fontsize=18)
plt.xticks(fontsize=12, rotation=90)
plt.yticks(fontsize=12, rotation=90)
plt.legend(title="Atomic number", fontsize=10)
# show graph
plt.show()
```
